### Funciones Recursivas
- Una funci贸n recursiva es una funci贸n que se llama a s铆 misma

Estructura de una funci贸n recursiva

    def funcion_recursiva(valor):

- def funcion_recursiva(valor): es la definici贸n de la funci贸n

- return valor es el valor a devolver

```python
#Ejemplo 12, Crear una funci贸n recursiva para obtener el 10 n煤mero par

print ("Ejemplo 12")
print ("1. Definir funci贸n")
def numero_par(numero):
    if numero == 0:
        return 0
    else:
        return numero_par(numero-1) + 2

print ("2. Llamar funci贸n")
resultado = numero_par(10)
print (resultado)

#Ejercicio 12, Crear una funci贸n recursiva para obtener el factorial de un n煤mero $5! = 5*4*3*2*1 = 120$

def factorial(numero):
    if numero == 0:
        return 1
    else:
        return numero*factorial(numero-1)

resultado = factorial(5)
print (resultado)
```
